### melihat missing value dari data tersebut, apabila terdapat missing value maka data perlu dibenahi atau tidak

```python
def chek_missing(df):
    sum_nan = df.isna().sum().reset_index()
    sum_nan.columns = ['kolom', 'jumlah_NaN']
    sum_nan["persentase"] = sum_nan.jumlah_NaN/len(df)*100
    return sum_nan.sort_values("persentase", ascending =  False)

dataset_missing = chek_missing(dataset)
test_missing = chek_missing(test)
train_missing = chek_missing(train)
```

```python
# outliers report
def outliers(s):
    iqr = (np.quantile(s, 0.75))-(np.quantile(s, 0.25))
    upper_bound = np.quantile(s, 0.75)+(1.5*iqr)
    lower_bound = np.quantile(s, 0.25)-(1.5*iqr)
    f = []
    for i in s:
        if i > upper_bound:
            f.append(i)
        elif i < lower_bound:
            f.append(i)
    sums = len(f)
    pros = len(f)/len(s)*100
    d = {'IQR':iqr,
         'Upper Bound':upper_bound,
        'Lower Bound':lower_bound,
        'Sum outliers': sums,'percentage outliers':pros}
    d = pd.DataFrame(d.items(),columns = ['sub','values'])
    return(d)
```

```python
# mencari outliers details
for i in test_numeric:
  print(i,'\n', display(outliers(test2[i])))
```

```python
# mencari outliers details
for i in train_numeric:
  print(i,'\n', display(outliers(train2[i])))
```

```python
def corr_plot(test2, metode):
  plt.figure(figsize=(30,15))
  triangle = np.triu(np.ones_like(test2.corr(method = metode)))
  heatmap = sns.heatmap(test2.corr(), vmin=-1, vmax=1, annot=True, mask = triangle)
  heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':15}, pad=12)
  return heatmap
```

```python
corr_plot(test2, "spearman");
```

```python
corr_plot(train2, "spearman");
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```
